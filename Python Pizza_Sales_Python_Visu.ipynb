### Hourly trend total orders

```python
# Conversion de la colonne 'order_time' en format datetime (heures, minutes, secondes)
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S')

# Extraction de l'heure (023)  partir de la colonne 'order_time'
df['order_hour'] = df['order_time'].dt.hour

# Regroupement des commandes par heure et comptage du nombre d'identifiants de commande uniques
orders_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique()

# Cration d'un graphique en barres reprsentant le nombre de commandes par heure
ax = orders_by_hour.plot(kind='bar', figsize=(10,5), color='maroon', edgecolor='black')

# Ajout d'un titre au graphique
plt.title("Total Orders by Hour of Day")

# tiquette de l'axe des abscisses (heures de la journe)
plt.xlabel("Hour of Day (24-Hour Format)")

# tiquette de l'axe des ordonnes (nombre de commandes)
plt.ylabel("Number of Orders")

# Rotation des tiquettes de l'axe des x  0 degr (horizontal)
plt.xticks(rotation=0)

# Ajout des valeurs exactes au-dessus de chaque barre
for i, val in enumerate(orders_by_hour):
    plt.text(i, val + 5, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

# Ajustement automatique de la mise en page pour viter les chevauchements
plt.tight_layout()

# Affichage du graphique
plt.show()
```

# Monthly trend total orders

```python
# Convertir la colonne 'order_date' en format datetime
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)

# Extraire le nom du jour de la semaine
df['month_name'] = df['order_date'].dt.month_name()

# Liste ordonne des mois
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Conversion de la colonne 'month_name' en variable catgorielle avec un ordre dfini
df['month_name'] = pd.Categorical(df['month_name'], categories=month_order, ordered=True)

# Regroupement du nombre de commandes uniques par mois
orders_by_month = df.groupby('month_name', observed=False)['order_id'].nunique()

# Cration d'une figure pour le graphique
plt.figure(figsize=(10,5))

# Zone colore sous la courbe pour amliorer la lisibilit
plt.fill_between(orders_by_month.index, orders_by_month.values, color='orange', alpha=0.6)

# Trac de la ligne principale reprsentant le nombre de commandes par mois
plt.plot(orders_by_month.index, orders_by_month.values, color='black', linewidth=2, marker='o')

# Ajout du titre et des tiquettes des axes
plt.title("Total Orders by Month")
plt.xlabel("Month")
plt.ylabel("Number of Orders")

# Rotation lgre des noms de mois pour une meilleure lisibilit
plt.xticks(rotation=45)

# Ajout des valeurs exactes au-dessus des points de donnes
for i, val in enumerate(orders_by_month):
    plt.text(i, val + 20, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

# Ajustement automatique de la mise en page pour viter les chevauchements
plt.tight_layout()

# Affichage du graphique
plt.show()

```

# % of Sales By Pizza Categories

```python
# Calcul du total des ventes par catgorie de pizza
category_sales = df.groupby('pizza_category')['total_price'].sum()

# Calcul du pourcentage des ventes pour chaque catgorie
category_pct = category_sales / category_sales.sum() * 100

# Dfinition de la taille du graphique
plt.figure(figsize=(5,5))

# Slection dune belle palette de couleurs prdfinie
colors = plt.get_cmap('tab20').colors  # nice color palette

# Cration du diagramme circulaire (camembert)
plt.pie(
    category_pct,
    labels=category_pct.index,            # noms des catgories
    autopct='%1.1f%%',                    # format daffichage des pourcentages
    startangle=90,                        # rotation de dpart pour le premier secteur
    colors=colors,                        # utilisation de la palette choisie
    wedgeprops={'edgecolor': 'black', 'width': 0.4}  # bordures et largeur des secteurs
)

# Ajout du titre au graphique
plt.title("Percentage of Sales by Pizza Category")

# Affichage du graphique
plt.show()
```

### % Of Sales Pizza By Size & Category

```python
# Cration d'un tableau crois dynamique (pivot table)  partir du DataFrame df
# Index : catgorie de pizza (ex. : Vgtarienne, Viande, etc.)
# Colonnes : taille de pizza (ex. : Small, Medium, Large)
# Valeurs : somme des prix totaux des ventes
# aggfunc='sum' : on additionne les ventes pour chaque combinaison catgorie/taille
# fill_value=0 : on remplace les valeurs manquantes par 0
sales_pivot = df.pivot_table(
    index='pizza_category',
    columns='pizza_size',
    values='total_price',
    aggfunc='sum',
    fill_value=0
)

# Calcul du pourcentage de chaque cellule par rapport au total des ventes
# Cela permet de visualiser la part relative de chaque catgorie/taille
sales_pct = sales_pivot / sales_pivot.sum().sum() * 100

# Dfinition de la taille de la figure pour la visualisation
plt.figure(figsize=(10,6))

# Cration d'une carte thermique (heatmap) avec seaborn
# annot=True : affiche les valeurs dans chaque cellule
# fmt='.1f' : format des nombres avec une dcimale
# cmap='YlOrRd' : palette de couleurs du jaune au rouge
# linewidths=0.5 : paisseur des lignes entre les cellules
sns.heatmap(sales_pct, annot=True, fmt='.1f', cmap='YlOrRd', linewidths=0.5)

# Ajout d'un titre au graphique
plt.title("% of Sales by Pizza Category and Size")

# tiquettes des axes
plt.xlabel("Pizza Size")
plt.ylabel("Pizza Category")

# Affichage du graphique
plt.show()
```

```python
# Regroupe les donnes par catgorie de pizza et calcule la somme des quantits vendues
pizzas_by_category = df.groupby('pizza_category')['quantity'].sum()

# Rcupre une liste de couleurs  partir de la palette 'tab20' de matplotlib
colors = list(plt.get_cmap('tab20').colors)

# Ajuste le nombre de couleurs  celui des catgories de pizza
colors = colors[:len(pizzas_by_category)]

# Cre un graphique en barres avec les quantits vendues par catgorie
# figsize=(8,5) : taille du graphique
# color=colors : couleurs personnalises pour chaque barre
# edgecolor='black' : bordure noire autour des barres
ax = pizzas_by_category.plot(kind='bar', figsize=(8,5), color=colors, edgecolor='black')

# Ajoute un titre au graphique
plt.title('Total Pizzas Sold by Pizza Category')

# tiquettes des axes
plt.xlabel('Pizza Category')
plt.ylabel('Pizzas Sold')

# Fait pivoter les noms des catgories sur laxe X pour une meilleure lisibilit
plt.xticks(rotation=45)

# Ajoute les valeurs au-dessus de chaque barre pour les rendre visibles
# enumerate(pizzas_by_category) : boucle sur chaque catgorie avec son index
# plt.text(...) : place le texte au-dessus de la barre
for i, val in enumerate(pizzas_by_category):
    plt.text(i, val + 5, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

# Ajuste automatiquement les marges pour viter que le texte soit coup
plt.tight_layout()

# Affiche le graphique
plt.show()
```
